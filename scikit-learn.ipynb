```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris

iris = load_iris()
```

```python
df= pd.DataFrame(
    data=np.c_[iris['data'], iris['target']],
    columns=iris['feature_names'] + ['target']
)

df.head()
```

```python
df.agg(['min', 'mean', 'median', 'max'])
```

```python
sns.scatterplot(x = df['sepal length (cm)'], y = df['petal length (cm)'], hue=df.target)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model = model.fit(df[['sepal length (cm)', 'petal length (cm)']], df['target'])

predicted_target = model.predict(df[['sepal length (cm)', 'petal length (cm)']])

sns.scatterplot(x = df['sepal length (cm)'], y = df['petal length (cm)'], hue=predicted_target == df['target'])
plt.show()
```
