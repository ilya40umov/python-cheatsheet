```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(934)

x = np.arange(0, 30)
y1 = x ** 2 + np.random.random(30) * x * 10
y2 = x ** 2 + 0.5 * x * 10

np.array([x[:5], y1[:5], y2[:5]])
```

```python
plt.scatter(x, y1, color='blue')
plt.plot(x, y2, color='red')
plt.show()
```

```python
fig = plt.figure()

sp1 = fig.add_subplot(3, 1, 1)
sp2 = fig.add_subplot(3, 1, 2)
sp3 = fig.add_subplot(3, 1, 3)

sp1.plot(x, y1, 'r--')
sp2.plot(x, y2, 'g-')
sp3.plot(x * 2, y2 - y1, 'b.')

plt.show()
```

```python
fig, sps = plt.subplots(3, 1, sharex=True)

sps[0].plot(x, y1, 'ro--')
sps[1].plot(x, y2, 'go-')
sps[2].plot(x * 2, y2 - y1, 'b.')

plt.show()
```

```python
plt.hist(y2 - y1, bins=15)

plt.show()
```

```python
data1 = [1, 2, 3, 4, 5]
data2 = [1, 2, 3, 2, 1]

plt.bar(np.arange(len(data1)), data1, width=0.3)
plt.bar(np.arange(len(data2)) + 0.3, data2, width=0.2)

plt.show()
```
